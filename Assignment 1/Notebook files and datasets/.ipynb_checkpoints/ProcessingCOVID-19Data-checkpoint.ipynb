```python
import pandas as pd
url = "https://dshs.texas.gov/coronavirus/TexasCOVID19DailyCountyCaseCountData.xlsx" #  sheets 'COVID-19 Cases', 'COVID-19 Fatalities'
df = pd.read_excel(url , sheet_name = 'Cases by County',skiprows=2)
df
```

```python
df=df.truncate(before=0,after=253)  # There are extra rows in the file, just remove, should look for last county, but this is easier
df
```

```python
df.keys()
```

```python
df2 = df.set_index('County Name')
df2
```

```python
from datetime import datetime
import re
pattern = re.compile('[0-9]+-[0-9]+', re.IGNORECASE)
dates = [datetime.strptime(pattern.findall(sub)[0], '%m-%d') for sub in df2.keys()]   
dates
```

```python
import re
df3 = df2
pattern = re.compile('[0-9]+-[0-9]+', re.IGNORECASE)
df3.columns=[pattern.findall(sub)[0] for sub in df3.keys()]
df3
```

```python
df2
```

```python
df3.T['Nueces'].plot()
```

```python
import matplotlib.pyplot as plt
plt.plot(df3.T['Nueces'])
```

```python
plt.plot(dates,df3.T['Nueces'])
```

```python
import matplotlib.dates as mdates
ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %d"))
plt.plot(dates,df3.T['Nueces'])
```

```python
mydf = df3.T.copy()

plt.figure()
ax=plt.gca()
ax.title.set_text('Nueces County Cases')
plt.plot(dates,mydf['Nueces'].diff(),label= 'Daily')
plt.plot(dates,mydf['Nueces'].diff().rolling(window=7).mean(),label= '7-day average')
ax.legend(ncol=3)
```

```python
plt.figure()
ax=plt.gca()
ax.title.set_text('Nueces County Cases')
plt.bar(dates,mydf['Nueces'].diff(),label= 'Daily')
plt.plot(dates,mydf['Nueces'].diff().rolling(window=7).mean(),label= '7-day average',color='red')
ax.legend()
```

```python
fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(1,2,2)
ax1.title.set_text('Nueces County Cases')
plt.bar(dates,mydf['Nueces'].diff(),label= 'Daily')
plt.plot(dates,mydf['Nueces'].diff().rolling(window=7).mean(),label= '7-day average',color='red')
ax1.legend()


ax2 = fig.add_subplot(1,2,1)
ax2.title.set_text('Dallas County Cases')
plt.bar(dates,mydf['Dallas'].diff(),label= 'Daily')
plt.plot(dates,mydf['Dallas'].diff().rolling(window=7).mean(),label= '7-day average',color='red')
ax.legend()


```
